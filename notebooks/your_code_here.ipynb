```python
import numpy as np
import pandas as pd
```

```python
#Challenge 1

movies = pd.read_csv('../data/input/IMDB-Movie-Data.csv')

def ratings(votes):
    if votes <= 999:
        return 'cat_1'
    elif votes <= 9999:
        return 'cat_2'
    elif votes <= 99999:
        return 'cat_3'
    elif votes <= 999999:
        return 'cat_4'
    else:
        return 'cat_5'
        
movies['Bin'] = movies['Rating'].apply(lambda column: ratings(column))
movies.head()
```

```python
#challenge 2

def revenue_pm(revenue, duration):
    return revenue / duration

movies['Revenue_per_minute'] = movies.apply(lambda row: revenue_pm(row['Revenue (Millions)'], row['Runtime (Minutes)']), axis=1)
movies
```

```python
#Challenge 3

def new_rating(genre, rating):
    if 'Thriller' in genre:
        return rating + 1
    elif 'Comedy' in genre:
        return rating - 1
    else:
        return rating
    
movies['New_rating'] = movies.apply(lambda row: new_rating(row['Genre'], row['Rating']), axis=1)
movies
```

```python
#Challenge 4

def ascii(title,votes):
    ascii_sum = sum(ord(char) for char in title)
    result = ascii_sum / votes
    return int(result)

def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
movies['Prime_number'] = movies.apply(lambda row: is_prime(ascii(row['Title'], row['Votes'])), axis=1)
movies
```
