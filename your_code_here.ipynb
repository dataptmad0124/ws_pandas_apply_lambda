# Challenge 1

```python
def categorize(value):
    if value < 1000:
        return 'cat1'
    elif value >= 1000 and value <10000:
        return 'cat2'
    elif value >= 10000 and value <100000:
        return 'cat3'
    elif value >= 100000 and value <1000000:
        return 'cat4'
    else:
        return 'cat5'
```

```python
 categorize(100000000)
```

```python
df.apply(lambda row: categorize(row['Votes']), axis=1)
```

```python
df.columns
```

```python
df['bins']=df.apply(lambda row: categorize(row['Votes']), axis=1)
```

```python
df
```

# Challenge 2. Using two arguments

```python
def rev_minute(value1, value2):
    if value1!=0:
           return value2/value1
   # elif value2 ==None:
      #  return 0    luego lo llenamos con FillNa
    else:
        return 0
```

```python
rev_minute(0,4)
```

```python
df['rev_minute']=df.apply(lambda row: rev_minute(row['Revenue (Millions)'], row['Runtime (Minutes)']), axis=1)
```

```python
df
```

# Challenge 3. A bit more complicated

```python
def new_rating(genre,rating):
    if  'Thriller'in genre and 'Comedy' not in genre:
        return rating+1
    elif 'Comedy' in genre and 'Thriller' not in genre:
        return rating -1
    
    else:
        return rating
   
```

```python
new_rating('ThrillerComedyj', 100) #prueba
```

```python
df['new_rating']=df.apply(lambda row:new_rating(row['Genre'], row['Rating']), axis=1)
```
